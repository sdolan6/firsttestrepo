```python
# Vignere Cipher ALTERNATE: give code and secret message you intend and 
# this will give the keyword that should be used to encode
# with HOW code and ARE actual (KEY: HXS): vig key H is a back 7 shift 

import string

code = input('CIPHER: ')
code = code.lower()
plaintext = input('SECRET MESSAGE: ')
plaintext = plaintext.lower()

alphabet = ['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z']


code = code.split(' ')
plaintext = plaintext.split(' ')
codenum = []
keywordsolve = []
wordcount = len(code)

for word in range (0,wordcount):
    wordind = 0
    eachcodeword = code[word]
    eachplainword = plaintext[word]
    eachcodeword = eachcodeword.translate(str.maketrans('','', string.punctuation))
    eachplainword = eachplainword.translate(str.maketrans('','', string.punctuation))
    for letter in eachcodeword:
        if letter.isalpha() == 1 and letter != "'":
            alphaind = alphabet.index(letter)
            plainind = eachplainword[wordind]
            plainind = alphabet.index(plainind)
            if alphaind > plainind:
                #back shift
                shift = alphaind - plainind
                keyletter = alphabet[shift]
                keywordsolve.append(keyletter)
            elif plainind > alphaind:
                #forward shift
                shift = plainind - alphaind
                keyletter = alphabet[-shift]
                keywordsolve.append(keyletter)
            else:
                keyletter = alphabet[0]
                keywordsolve.append(keyletter)
            wordind = wordind + 1

keywordsolve = ''.join(keywordsolve)
keywordsolve = keywordsolve.upper()

keywordsolve
```
